```python
#Import pandas 
import pandas as pd 


df = pd.read_csv("mp_data1.csv")
df.head(30)
```

```python
df1 = df[df['sender'] == 'target']
df1.head()
```

```python
set(df1['sender'])
```

```python
#in order to make the 'correct'-column to state the right thing (right now it doesn't), we need to 
#change the 'response_action'-column to only contain n or m's
df1['response_action'] = df1['response_action'].str.replace('keypress(', '')
df1['response_action'] = df1['response_action'].str.replace(')', '')
df1.head()
```

```python
df1['correct'] = df1['response_action'] == df1['correct_response']
df1.head()
```

```python
set(df1['correct'])
```

```python
list(df1)
```

```python
#filter only the columns of interest

data = df1.loc[:, ['ID', 'duration', 'prime', 'condition', 'soa_condition', 'congruence', 'correct']]
data.head()
```

```python
# filter the practice rounds out
data = data[data['condition'] != 'practice']
data.head()
```

```python
#now we have a dataframe containing only the information that we need. Time for plotting the data
import seaborn as sns

plt = sns.pointplot(data=data, x='soa_condition', y='duration', hue='congruence', order = ['short', 'long'])
plt.set_ylabel('RT (ms)')
plt.set_xlabel('SOA Condition')
plt.title('My awesome plot')
```
