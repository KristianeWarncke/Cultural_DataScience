## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
```

```python
#Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
df1 = rawdata[0]
df2 = rawdata[3]
df3 = pd.concat([df1,df2])
df3
```

```python
#Adds a column which marks whether each row is from the list of physical singles or digital singles
#En måde at gøre det på
df3['Format'] = pd.cut(x = df3['Released'], bins = [1900,2000,2024], labels = ['physical', 'digital'])
df3
```

```python
#Adds a column which marks whether each row is from the list of physical singles or digital singles
#en anden måde at gøre det på
Format = []

for idx, row in df3.iterrows():
    if row['Single'] in df1['Single'].values:
        Type = "physical"
    elif row['Single'] in df2['Single'].values:
        Type = "digital"
    Format.append(Type)

df3['Format'] = Format

df3
```

```python
# Sorting the dataframe by Released year
df3 = df3.sort_values(by=['Released'])
df3
```

```python
# Check if the year 2000 is present
if 2000 in df3['Released'].values:
    # If 2000 is found, filter by the exact year
    result = df3[df3['Released'] == 2000]
else:
    # If 2000 is not found, show the closest year after 2000
    result = df3[df3['Released'] > 2000].head(1)
# Print the artist and single for the matching year or closest year after
    print("The single released closest after the year I was born is", result['Artist'].values[0], ",", result['Single'].values[0], ",", result['Released'].values[0])
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
#Make a single dataframe that combines the space missions from the 1950's to the 2020's'
df4 = pd.concat(rawdata[:8])
df4
```

```python
#Write a script that returns the year with the most launches

# Extracting year from Launch date
df4['Year'] = pd.to_datetime(df4['Launch date']).dt.year
df4

# Count occurrences of each year
year_counts = df4['Year'].value_counts()

# Find the year with the most launches
most_launches_year = year_counts.idxmax()
most_launches_count = year_counts.max()

# Print the year with the most launches and the count
print(f"The year with the most launches is {most_launches_year} with {most_launches_count} launches.")
```

```python
#Write a script that returns the most common month for launches

#extract months from launch date column
df4['Month'] = pd.to_datetime(df4['Launch date']).dt.month
df4

# Count occurrences of each month
month_counts = df4['Month'].value_counts()


# Find the month with the most launches
most_launches_month = month_counts.idxmax()
most_launches_counts = month_counts.max()

# Print the year with the most launches and the count
print(f"The month with the most launches is month no. {most_launches_month} with {most_launches_counts} launches.")
```

```python
month_counts
```

```python
#Write a script that ranks the months from most launches to fewest launches

# Convert the month_counts to a DataFrame
month_counts_df = month_counts.reset_index()

# Rename the columns for clarity
month_counts_df.columns = ['Month', 'Launch Count']

# Merge with the original DataFrame based on the 'Month' column
df4 = df4.merge(month_counts_df, on='Month', how='left')
df4 = df4.sort_values(by=["Launch Count"])
df4
```

## Supervillain challenge

1. Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
2. Write a script that ranks each decade in terms of how many supervillains debuted in that decade
3. Write a script that ranks the different comics companies in terms of how many supervillains they have, and display the results in a nice table (pandas dataframe)

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
#Write a script that combines the tables showing supervillain debuts from the 30's through the 2010's
df5 = pd.concat(rawdata[3:11])
df5
```

```python
#Write a script that ranks each decade in terms of how many supervillains debuted in that decade

#Extract the year from 'Year Debuted' column (ignoring parentheses content)
df5['Year'] = df5['Year Debuted'].str.extract(r'(\d{4})').astype(int)

#Convert the year to the corresponding decade
df5['Decade'] = (df5['Year'] // 10) * 10

#Group by the 'Decade' and count the number of rows (debuts) per decade
debuts_per_decade = df5.groupby('Decade').size().reset_index(name='Count')

#Sort by the count in descending order (most debuts at the top)
debuts_per_decade = debuts_per_decade.sort_values(by='Count', ascending=False).reset_index(drop=True)

debuts_per_decade
```

```python
#### Write a script that ranks the different comics companies in terms of how many supervillains they have, 
#and display the results in a nice table (pandas dataframe)

set(df5['Company'].values)
```

```python
#Group by the 'company' and count the number of rows (debuts) per company
debuts_per_company = df5.groupby('Company').size().reset_index(name='Count')

#Sort by the count in descending order (most debuts at the top)
debuts_per_company = debuts_per_company.sort_values(by='Count', ascending=False).reset_index(drop=True)

debuts_per_company
```
